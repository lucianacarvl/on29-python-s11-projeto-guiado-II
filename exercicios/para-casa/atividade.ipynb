```python

def criar_db():
    conn = sqlite3.connect("Song.db")
    cursor = conn.cursor()
    cursor.execute("""CREATE TABLE IF NOT EXISTS Song(
        Artist VARCHAR NOT NULL,
        Title VARCHAR NOT NULL,
        Year INT NOT NULL,
        Sales FLOAT NOT NULL,
        Streams FLOAT NOT NULL,
        Downloads FLOAT NOT NULL,
        RadioPlays FLOAT NOT NULL,
        Rating FLOAT);
""")
criar_db()
```

```python
def importar_csv():
    conn = sqlite3.connect("Song.db")
    cursor = conn.cursor()
    
    with open('Song.csv', newline='', encoding='utf-8') as csvfile:
        leitor = csv.reader(csvfile)
        next(leitor)
        for linha in leitor:
            cursor.execute("INSERT INTO Song (Artist, Title, Year, Sales, Streams, Downloads, RadioPlays, Rating) VALUES (?, ?, ?, ?, ?, ?, ?, ?)", (linha[0], linha[1], linha[2], linha[3], linha[4], linha[5], linha[6], linha[7]))
            cursor.execute("SELECT * FROM Song")
    conn.commit()
    cursor.close()
    conn.close()
importar_csv()
```
